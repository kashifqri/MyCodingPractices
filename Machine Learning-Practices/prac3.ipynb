```python
q1=datasets["id"].quantile(0.25)
q1
```

```python
q3=datasets["id"].quantile(0.75)
q3
```

```python
IQR=q3-q1
```

```python
# min_range=q1-(1.5*IQR) # using IQR function
min_range=datasets["id"].mean()-(3*datasets["id"]).std() #using Z score first methord

# max_range=q1+(1.5*IQR) # using IQR function
max_range=datasets["id"].mean()+(3*datasets["id"]).std()#using Z score first methord
min_range,max_range
```

```python
newdataset=datasets[datasets["id"]<=max_range]
newdataset
```

```python
newdataset.shape
```

```python
sns.boxplot(x="id",data=newdataset)
plt.show()
```

```python
z_score=(datasets["id"]-datasets["id"].mean())/(datasets["id"].std()) #second methord of Z score
z_score
```

```python
datasets["z_score"]=z_score
```
