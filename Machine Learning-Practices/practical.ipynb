```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
var=sns.load_dataset("titanic")
var
```

```python
x = var.drop(columns=["sex", "embarked", "class", "who", "adult_male", "deck", "embark_town", "alive", "alone"], axis=1).head(20)
x
```

```python
sns.heatmap(x,annot=True,)
plt.show()
```

```python
var=sns.load_dataset("titanic").head(20)
var
```

```python
sns.pairplot(var,hue="sex",kind="kde")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(y="fare", x="sex", data=var, hue="sex", palette="dark")
plt.show()
```

```python
sns.catplot(y="fare", x="sex", data=var, hue="sex", palette="Oranges",kind="boxen")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.catplot(y="fare", x="sex", data=var, hue="sex", palette="Oranges",kind="boxen")
sns.despine()
sns.set_context("poster",font_scale=10)
plt.show()
```

```python
fg = sns.FacetGrid(var,col="sex")  # Adjust size with height and aspect
fg.map(plt.scatter, "fare", "adult_male")
plt.tight_layout()
plt.show()



```
