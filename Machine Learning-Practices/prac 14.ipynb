```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
datasets=pd.read_csv(r"C://Users//basit//Downloads//Iris.csv")
datasets.head(3)
```

```python
datasets.drop("Species",axis=1,inplace=True)
```

```python
datasets
```

```python
sns.pairplot(data=datasets)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range (2,21):
    km=KMeans(n_clusters=i,init="k-means++")
    km.fit(datasets)
    wcss.append(km.inertia_)
```

```python
plt.plot([i for i in range (2,21)],wcss,marker="o")
plt.xlabel("no of curse")
plt.xticks([i for i in range (2,21)])
plt.ylabel("wcss")
plt.show
```
