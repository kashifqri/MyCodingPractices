```python
from sklearn.impute import SimpleImputer
```

```python
print(f"the number of missings valus in before imputing :{df.age.isnull().sum()}")
```

```python
imp = SimpleImputer(strategy="mean")
df["age"] = imp.fit_transform(df[["age"]])

print(f"The number of missing values after imputing: {df.age.isnull().sum()}")
```

```python
def get_parameters(df):
    parameters = {}
    for col in df.columns[df.isnull().any()]:  # Loop through columns with missing values
        if df[col].dtype in ["float64", "int64", "int32"]:
            strategy = "mean"
        else:
            strategy = "most frequent"
        
        missing_values = df[col].isnull().sum()  # Count of missing values in the column
        parameters[col] = {"missing_val": missing_values, "strategy": strategy}
    
    return parameters
    
get_parameters (df) 
    
```

```python
parameters=get_parameters(df)
```

```python
for col, param in parameters.items():  
    missing_values = param['missing_val']
    strategy = param["strategy"]
    imp = SimpleImputer(missing_values= missing_values, strategy = strategy)
    df[col] = imp.fit_transform(df[[col]])

```
