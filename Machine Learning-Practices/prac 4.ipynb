```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
datasets=pd.read_excel("C://Users//basit//Desktop//Book1.xlsx")
datasets
```

```python
datasets.isnull().sum()
```

```python
datasets.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss= StandardScaler()
ss.fit(datasets[["id"]])
```

```python
id_1=ss.transform(datasets[["id"]])
```

```python
datasets["id_1"]=id_1
```

```python
datasets
```

```python
datasets.describe()
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1,2,1)
plt.title("before")
sns.distplot(datasets["id"])

plt.subplot(1,2,2)
plt.title("after")
sns.distplot(datasets["id_1"])

plt.show()
```
