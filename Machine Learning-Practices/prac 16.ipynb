```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
import pandas as pd
```

```python
x, y = make_moons(n_samples=250, noise=0.05)
```

```python
df = { "data1":x[:,0],"data2":x[:,1]}
```

```python
datasets=pd.DataFrame(df)
```

```python
datasets
```

```python
sns.scatterplot(x="data1",y="data2",data=datasets)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db=DBSCAN(eps=0.2, min_samples=5)
datasets["predict"]=db.fit_predict(datasets)
```

```python
sns.scatterplot(x="data1",y="data2",data=datasets,hue="predict")
plt.show()
```
