```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
```

```python
datasets=pd.read_csv(r"C://Users//basit//Downloads/placement.csv")
datasets
```

```python
sns.scatterplot(x="cgpa",y="placement",data=datasets)
plt.show()
```

```python
x=datasets[["cgpa"]]
y=datasets["placement"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import SVR
```

```python
sm=SVR(kernel="poly")
sm.fit(x_train,y_train)
```

```python
sm.score(x_test,y_test)
```

```python
sm.score(x_train,y_train)
```

```python
sns.scatterplot(x="cgpa",y="placement",data=datasets)
plt.scatter(datasets["cgpa"],sm.predict(x),color="red")
plt.show()
```
