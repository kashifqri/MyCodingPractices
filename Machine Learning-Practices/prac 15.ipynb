```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
datasets=pd.read_csv(r"C://Users//basit//Downloads//Iris.csv")
datasets.head(3)
```

```python
datasets.drop("Species",axis=1,inplace=True)
```

```python
sns.pairplot(data=datasets)
plt.show()
```

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20,20))
sc.dendrogram(sc.linkage(datasets, method='single', metric='euclidean'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 
```

```python
ac=AgglomerativeClustering(n_clusters=2,linkage="single")
datasets["predict"]=ac.fit_predict(datasets)
```

```python
datasets
```

```python
sns.pairplot(data=datasets,hue="predict")
plt.show()
```
