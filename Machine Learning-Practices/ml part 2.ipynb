```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_decision_regions
```

```python
datasets=pd.read_csv(r"C://Users//basit//Downloads//Social_Network_Ads.csv")
datasets
```

```python
datasets.isnull().sum()
```

```python
x=datasets.iloc[:,:-1]
y=datasets["Purchased"]
```

```python
sc=StandardScaler()
sc.fit(x)
x=pd.DataFrame(sc.transform(x),columns=x.columns)
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
ds=DecisionTreeClassifier(max_depth=2)
ds.fit(x_train,y_train)      #ds.fit(x.iloc[:, :-1].to_numpy(), y.to_numpy())
```

```python
plot_decision_regions(x.iloc[:, :-1].to_numpy(), y.to_numpy(), clf=ds)
plt.show()
```

```python
ds.score(x_test,y_test)*100
```

```python
ds.score(x_train,y_train)*100
```
